```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('IAA-image/RepScores/RepScore1.csv', header = None)
data
```

```python
data.loc[data[1] == 1, 4]
```

```python
# Users 1 - 3 consistently agree with 0.9 agreement, user 4 starts at 0.5, increases to 0.9, then starts
# trolling immediately and decreases to 0.1. User 5 starts at 0.1 and increases by 0.1 each run until at 
# 0.9
plt.figure(figsize = (16, 9))
plt.plot(range(10), data.loc[data[1] == 1, 4], label = 1)
plt.plot(range(10), data.loc[data[1] == 2, 4], label = 2)
plt.plot(range(10), data.loc[data[1] == 3, 4], label = 3)
plt.plot(range(10), data.loc[data[1] == 4, 4], label = 4)
plt.plot(range(10), data.loc[data[1] == 5, 4], label = 5)
plt.legend()
plt.show()
```

```python
data2 = pd.read_csv('IAA-image/RepScores/RepScore2.csv', header = None)
data2
```

```python
#User 1 constantly increases agreement by 0.01 each run until it reaches 1 (starting from 0.5). 
#Other users remain constant at 0.7
plt.figure(figsize = (16, 9))
plt.plot(range(50), data2.loc[data2[1] == 1, 4], label = 1)
plt.plot(range(50), data2.loc[data2[1] == 2, 4], label = 2)
plt.plot(range(50), data2.loc[data2[1] == 3, 4], label = 3)
plt.plot(range(50), data2.loc[data2[1] == 4, 4], label = 4)
plt.plot(range(50), data2.loc[data2[1] == 5, 4], label = 5)
plt.axvline(x=3, color = 'red')
plt.grid('on')
plt.legend()
plt.show()
```
