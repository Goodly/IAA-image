```python
data_runs = pd.DataFrame()
for i in range(50):
    d = pd.read_csv('Series_Output/Run' + str(i) + '.csv')
    data_runs = data_runs.append(d)
data_runs
```

```python
runs_data1 = data_runs.loc[data_runs['contributor_id'] == 1, 'answer_number']
runs_data2 = data_runs.loc[data_runs['contributor_id'] == 2, 'answer_number']
runs_data3 = data_runs.loc[data_runs['contributor_id'] == 3, 'answer_number']
runs_data4 = data_runs.loc[data_runs['contributor_id'] == 4, 'answer_number']
runs_data5 = data_runs.loc[data_runs['contributor_id'] == 5, 'answer_number']
runs_data_mean = pd.Series()

for i in range(500):
    runs_data_mean = np.append(runs_data_mean, (np.mean(np.array([runs_data1.iloc[i], runs_data2.iloc[i], 
                                                   runs_data3.iloc[i], runs_data4.iloc[i],
                                                  runs_data5.iloc[i]]))))
```

```python
runs_data1_diff = abs(runs_data1 - runs_data_mean)
runs_data2_diff = abs(runs_data2 - runs_data_mean)
runs_data3_diff = abs(runs_data3 - runs_data_mean)
runs_data4_diff = abs(runs_data4 - runs_data_mean)
runs_data5_diff = abs(runs_data5 - runs_data_mean)
```
